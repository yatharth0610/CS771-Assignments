## Part 1(a) and 1(b)

The below class implements the perceptron algorithm (fit function) as well as contains the functions required to make predictions (predict) using the algorithm for the banknotes dataset. 

```python
## Coding custom perceptron class

class Perceptron:
    
    ## Constructor
    def __init__(self):
        self.w = None
       
    ## Computing the expected output
    def y_hat(self, x):
        return 1 if (np.dot(self.w, x) >= 0) else -1

    ## Computing predictions
    def predict(self, X):
        result = []
        for x in X:
            result.append(self.y_hat(x))
            
        return np.array(result)
            
    ## Computes current loss using the 0-1 loss function
    def get_loss(self, X, Y):
        res = 0
        for i, row in enumerate(X):
            diff = 0
            if self.y_hat(row) != Y[i]:
                diff = 1
            res = res + diff 
        return res/X.shape[0]
    
    ## Function to fit the moodel
    def fit(self, X, Y, epochs = 10000, eta = 1):
        np.random.seed(10)
        self.w = np.random.randint(0, 10, X.shape[1])
        losses = [] ## Array to store loss for checking convergence
        
        for i in range(epochs):
            ## Choosing a random data point in every iteration
            idx = np.random.choice(X.shape[0])
            y_i = Y[idx]
            x_i = X[idx]
            
            ## Updating the weigths
            if y_i * (np.dot(self.w, x_i)) < 0:
                self.w = self.w + eta * y_i * x_i
            
            curr_loss = self.get_loss(X, Y)
            losses.append(curr_loss)
            
        return np.array(losses)
```

Now, we will implement the 5-fold cross validation to obtain the best model for this dataset. For this, we will first remove $20\%$ of the data as test data and perform 5-fold CV on the remaining. We will be cross validating for the parameter of learning rate and find which one fits the best.

```python
## Custom function that performs the split 
def cvSplit(df, folds):
    
    np.random.seed(1)
    
    ## Array that stores the split
    split = list()
    df = df.sample(frac = 1)
    df_cpy = df.copy(deep = True)
    sz = int(df_cpy.shape[0]/folds)
    
    for i in range(folds):
        fold = list()
        while len(fold) < sz:
            ## Generating random rows of dataset (Shuffling)
            idx = np.random.randint(df_cpy.shape[0])
            idx = df_cpy.index[idx]
            fold.append(df_cpy.loc[idx].values.tolist())
            df_cpy.drop(idx, inplace = True)
            
        split.append(np.array(fold))
            
    return np.array(split)

def CV(df, split, folds, lr):
    ## Array to store the losses 
    losses = list()
    
    for fold in range(folds):
        split_cpy = np.copy(split)
        print("Currently doing fold number " + str(fold + 1))
        
        ## Generating the test set for this iteration
        test_data = split_cpy[fold]
        X_test = test_data[:, :-1]
        Y_test = test_data[:, -1]
        
        ## Remaining data is train set
        train_data = np.delete(split_cpy, fold, axis = 0).reshape(-1, 5)
        X_train = train_data[:, :-1]
        Y_train = train_data[:, -1]
        
        ## Appending column of ones to accomodate for bias term
        dummy_train = np.ones((X_train.shape[0],1))
        dummy_test = np.ones((X_test.shape[0], 1))
        X_train = np.hstack((X_train, dummy_train))
        X_test = np.hstack((X_test, dummy_test))
        
        ## Changing labels from 0/1 to -1/1
        Y_test[Y_test == 0] = -1
        Y_train[Y_train == 0] = -1
        
        ## Training the model
        perc = Perceptron()
        _ = perc.fit(np.array(X_train), np.array(Y_train), 10000, lr)
        losses.append(perc.get_loss(np.array(X_test), np.array(Y_test)))
        
    losses = np.array(losses)
    
    ## Taking mean of errors for k-folds
    return np.mean(losses)
```

```python
## Choosing the best learning rate using 5-fold cross validation

## Generating splits
np.random.seed(10)
df = df.sample(frac = 1)

## Breaking dataset into 80:20 ratio for generating held-out set
sz = int(0.8 * df.shape[0])
df_train = df.iloc[:sz, :]
df_test = df.iloc[sz:, :]

print(df_train.shape, df_test.shape)
split = cvSplit(df_train, 5)

minn = 1
min_lr = -1

for i in range(1, 11, 1):
    print("Current value of lr " + str(i / 10))
    
    ## Obtaining loss from 5-fold CV and generated splits
    loss = CV(df_train, split, 5, i/10)
    print(loss)
    
    ## Finding the minimum error obtained
    if loss < minn:
        minn = loss
        min_lr = i/10
        
print("Minimum Loss Achieved " + str(minn) + " with learning rate of " + str(min_lr))
```

In this part, we will find out the f1 score on the held-out dataset we created before, using the best model we obtained with the learning rate of 0.6 (obtained using cross validation). 

```python
## Function to calculate the F1 score

def f1_score(pred, truth, lab):
    TP = np.sum((pred == truth) & (truth == lab))
    FN = np.sum((pred != truth) & (pred != lab))
    FP = np.sum((pred != truth) & (pred == lab))
    
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    return 2 * (precision * recall) / (precision + recall)
```

### Part 2(a)

In this part, we were asked to obtain the posterior distribution for a random variable, given it's prior and 50 observations from the true dataset. For this problem, we will assume that the random variable is normally distributed with variance already known to us, and given prior for the mean of the guassian.

Formally, we can define the problem as given a prior for $ \mu $ and given data from a distribution $N(\mu, \sigma)$ (with $\sigma$ known to us). 

$$X \sim N(\mu, \sigma)$$
$$\mu \sim N(\mu_{0}, \sigma_{0})$$

Now, given 50 samples of $X$ from true sample, we can update our prior belief about $\mu$ using bayes rule. We write this as 

$$P(\mu | \textbf{x}) = \frac{P(\textbf{x} | \mu)\cdot P(\mu)}{P(\textbf{x})}$$

Since, in our case the prior and likelihood both are gaussian so posterior will also be a gaussian (say with mean $\mu'$ and sd $\sigma'$). Since, the denominator term in the above equation will just be a constant term we can write the above equation as

$$c_{1}e^{\frac{-(\mu - \mu')^2}{2 \sigma'^2}} = c_{2}e^{\sum_{i = 1}^{50} \frac{-(x_{i} - \mu)^2}{2 \sigma^2}} \cdot e^{\frac{-(\mu - \mu_{0})^2}{2 \sigma_{0}^2}}$$

Ignoring the constants and matching the coefficients of $\mu^2$ from inside the exponential term, we obtain

$$\frac{1}{\sigma'^2} = \frac{1}{\sigma_{0}^2} + \frac{N}{\sigma^2}$$

Matching the coefficients of $\mu$ from inside the exponential term, we obtain

$$\mu' = \left( \frac{\mu_{0}}{\sigma_{0}^2} + \frac{\sum_{i = 0}^{50} x_{i}}{\sigma^2}\right) \cdot {\sigma'^2}$$

We can plug the value of $\sigma'$ obtained from the previous equation and update $\mu'$. In this way we will obtain the posterior distribution, which has been implemented in the cell below.

```python
## Updating the prior on mean after seeing the observations

def gauss_posterior(samp, N, mu_0, var_0, var_1): 
    var = 1 / ((1 / var_0) + (N / var_1))
    mu = (mu_0 / var_0 + sum(samp) / var_1) * var
    
    return (mu, np.sqrt(var))


N = 50
mu_0 = 25
var_0 = 25
var_1 = 25

mu, sigma = gauss_posterior(samp, N, mu_0, var_0, var_1)
print(mu, sigma)
```

Now, we will calculate the distribution of the likelihood. Since for any $\mu$, the likelihood function for that is a product of probabilities obtained from a guassian, we will have the likelihood to be normally distributed as well. So, we will try to obtain the mean and standard deviation for that guassian. 

Again proceeding as before we will assume that likelihood has mean $\mu'$ and variance $\sigma'$. Likelihood can be simply obtained by multiplying the probabilities of all the $x_{i}$ assuming that they have been sampled independently. Therefore, likelihood will just be a product of gusassians. And since we don't care about the constant terms (since we are asked to plot the pdf and hence the constant term will be unique), we will get

$$c_{1}e^{\frac{-(\mu - \mu')^2}{2 \sigma'^2}} = c_{2}e^{\sum_{i = 1}^{50} \frac{-(x_{i} - \mu)^2}{2 \sigma^2}} \text{with } c_2 = \left(\frac{1}{\sqrt{2\pi \sigma^2}}\right)^{50}$$

From this comparing the coefficients of $\mu^2$ inside exponent, we get

$$\frac{1}{\sigma'^2} = \frac{N}{\sigma^2}$$

And comparing the coefficients of $\mu$, we get

$$\mu' = \frac{\sum_{i = 0}^{50} x_i}{N}$$

I have used the above two results to obtain the function below. 

```python
## Function to obtain parameters of likelihood
def norm(x, mu, sigma):
    return (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu)**2 / (2 * sigma**2))

## Function to return the likelihood of obtaining a sample from a guassian
def likelihood(samp, mu, sigma):
    n = len(samp)
    ans = 1
    for i in range(n):
        ans = ans * norm(samp[i], mu, sigma)
    return ans

## Function to obtain parameters of likelihood
def gauss_likelihood(samp, N, var_0):
    var = var_0 / N
    mu = np.mean(samp)
    return (mu, np.sqrt(var))
```

## Part 2(b)

In this part we were asked to implement the Metropolis Algorithm and compare the posterior obtained from this with the original one. Since, in this problem we were asked to choose any proposal width, I have chosen a sd of 1. 

```python
### Implementing the Metropolis-Hasting algorithm

def metropolis_hastings(samp, N, sd, iterations, mu_0, var_0, var_1):
    np.random.seed(0)
    theta_0 = 0
    theta_1 = np.random.normal(theta_0, sd, 1)
    mu_1, sigma = gauss_likelihood(samp, N, var_1)
    r1 = norm(theta_1, mu_0, np.sqrt(var_0)) * norm(theta_1, mu_1, sigma)
    r2 = norm(theta_0, mu_0, np.sqrt(var_0)) * norm(theta_0, mu_1, sigma)
    
    r = r1/r2
    a = np.random.uniform(0,1)
    accepted = []
    if r > a:
        theta_0 = theta_1
        accepted.append(theta_0)
    
    for i in range(iterations):
        theta_1 = np.random.normal(theta_0, sd, 1)
        r1 = norm(theta_1, mu_0, np.sqrt(var_0)) * norm(theta_1, mu_1, sigma)
        r2 = norm(theta_0, mu_0, np.sqrt(var_0)) * norm(theta_0, mu_1, sigma)
    
        r = r1/r2
        a = np.random.uniform(0,1)
        if r > a:
            theta_0 = theta_1
            accepted.append(theta_0)
        
    return np.array(accepted)
```

```python
posterior = metropolis_hastings(samp, 50, 1, 300000, 25, 25, 25)
count, bins, ignored = plt.hist(posterior, 100, density = True)

## Actual posterior
mu, sigma = gauss_posterior(samp, 50, 25, 25, 25)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2)), color="r", label = "actual")
plt.legend(loc = "upper left")
plt.show()
```

As can be seen above, performing a large number of iterations (3e5), we get that the obtained draws match with the required/actual posterior, we were trying to estimate. We will see below that the proposal width plays a major role in the rate of convergence as we will see below.
